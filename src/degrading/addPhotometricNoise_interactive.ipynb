### Imports

```python
from rail.core.data  import DataStore
from rail.core.stage import RailStage
from rail.core.data  import PqHandle

from MultiSurveyErrorModel import MultiSurveyErrorModel
```

### Specify path and parameters

```python
noiseless_catalog_path = '/Users/leo/Projects/LBG_cosmology/simulated_catalogs/integrated_catalog_23apr26.pq'

### 5-sigma limiting depths
m5_limiting_depths = {'LSST_u'    : 27.74,
                      'LSST_g'    : 28.69,
                      'LSST_r'    : 28.88,
                      'LSST_i'    : 28.96,
                      'LSST_z'    : 28.26,
                      'LSST_y'    : 26.63,
                      'Roman_F062': 27.7,
                      'Roman_F087': 27.7,
                      'Roman_F106': 27.6,
                      'Roman_F129': 27.5,
                      'Roman_F158': 27.0,
                      'Roman_F184': 25.9,
                      'Roman_F213': 28.3,
                      'HSC_MB_00' : 26.41,
                      'HSC_MB_01' : 26.51,
                      'HSC_MB_02' : 26.45,
                      'HSC_MB_03' : 26.69,
                      'HSC_MB_04' : 26.93,
                      'HSC_MB_05' : 26.62,
                      'HSC_MB_06' : 26.26,
                      'HSC_MB_07' : 26.02,
                      'HSC_MB_08' : 26.07,
                      'HSC_MB_09' : 26.00,
                      'HSC_MB_10' : 26.06,
                      'HSC_MB_11' : 25.52,
                      'HSC_MB_12' : 25.58,
                      'HSC_MB_13' : 25.43,
                      'HSC_MB_14' : 25.15,
                      'HSC_MB_15' : 24.79}

bands = list(m5_limiting_depths.keys())

nYrObs     = 1
nVisYr     = 1
gamma      = 0.04
sigLim     = 0
inputType  = 'pogson'
outputType = 'asinh'
seed       = 42
```

### Set up DataStore and load catalog

```python
RailStage.data_store = DataStore()

RailStage.data_store['noiseless_catalog'] = PqHandle('noiseless_catalog', path=noiseless_catalog_path)
```

### Instantiate and run stage

```python
stage = MultiSurveyErrorModel.make_stage(

    name = 'MultiSurveyErrorModel',

    inputType  = inputType,
    outputType = outputType,

    m5     = m5_limiting_depths,
    bands  = bands,
    nYrObs = nYrObs,
    nVisYr = nVisYr,
    gamma  = gamma,
    sigLim = sigLim,

    seed = seed)

stage.run()
```

### Access result in memory

```python
noisy_catalog = stage.get_data('noisy_catalog')
noisy_catalog
```
